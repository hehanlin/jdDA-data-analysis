## 著名品牌

```python
brand_df = pd.DataFrame(good_list['brand_list'])
```

```python
brand_df.title.tolist()
```

```python
top_goods_df = pd.DataFrame(good_list['top_good_list'])
```

```python
def apply_commit_num(x):
    if x.find("万")!=-1:
        tmp = x.replace("万", "").replace("+", "")
        tmp = float(tmp)*10000
    elif x.find("+")!=-1:
        tmp = x.replace("+", "")
        tmp = float(tmp)
    else:
        tmp = float(tmp)
    return tmp
```

```python
top_goods_df['commit_num'] = top_goods_df['commit_num'].apply(apply_commit_num)
```

```python
top_goods_df
```

## top商品销量排名

```python
top_goods_df.sort_values("commit_num", ascending=False)
```

```python
bar = pyecharts.Bar("top商品销量排名")
```

```python
bar.add("销量", top_goods_df['title'], top_goods_df['commit_num'])
```

```python
bar
```

```python
top_goods_df = top_goods_df[top_goods_df['shop_name'].fillna("").map(lambda x: True if x.find("京东")!=-1 else False)]
```

```python
def replace_brand_name(x):
    x = x.replace("京东", "").replace("旗舰", "").replace("自营", "").replace("官方", "").replace("专营", "").replace("店", "")
    return x
```

```python
top_goods_df['brand_name'] = top_goods_df['shop_name'].apply(replace_brand_name)
```

```python
top_goods_df
```

```python
shop_num = top_goods_df['commit_num'].groupby(top_goods_df['brand_name']).sum()
```

## 品牌销量

```python
top_goods_df['price'] = top_goods_df['price'].fillna(0).astype(float)
```

```python
top_goods_df['total_price'] = top_goods_df['commit_num']*top_goods_df['price']
```

```python
total_price = top_goods_df['total_price'].groupby(top_goods_df['brand_name']).sum()
```

```python
bar1 = pyecharts.Bar("各品牌销售总额")
```
